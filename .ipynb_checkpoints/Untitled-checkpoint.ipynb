```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
W=H=400
bg_clr = (255,255,255)
vp = np.full((H, W, 3), bg_clr, dtype = 'uint8')

vp[:, :] = (255,255,255) #изменение всех элементов массива в одной строчке

# границы 1
vp[0:10] = (255,0,0)
vp[390:400] = (255,0,0)
vp[:,0:10]= (255,0,0)
vp[:,390: 400] = (255,0,0) 

# границы 2
width = 5
red = (255,0,0)
vp[:width,:] = red
vp[:,-width:] = red
vp[-width:,:] = red
vp[:,:width] = red
```

```python
#Функции для направлений
def right(vp,a,b,x,y,rgb,i,n):
    while(i<b):
        vp[y-i:y+i ,x+n:x+b] = rgb
        i+=1
        n=i*i
def left(vp,a,b,x,y,rgb,i,n):
    while(i<b):
        vp[y-i:y+i ,-x-b:-x-n] = rgb
        i+=1
        n=i*i
def up(vp,a,b,x,y,rgb,i,n):
    while(i<a):
        vp[-y-b:-y-n ,x-i:x+i ] = rgb
        i+=1
        n=i*i
        
def down(vp,a,b,x,y,rgb,i,n):
    while(i<b):
        vp[x+n:x+b,y-i:y+i] = rgb
        i+=1
        n=i*i

def Zacrash(k,j,m,vp,a,b,x,y,bg_clr,i,n):
    rgb = np.random. randint(0,255, size  =3)
    if k==0:
        if np.all( vp[y-a:y+a , x:x+b ] == bg_clr) :
            right(vp,a,b,x,y,rgb,i,n)
            j+=1
        m+=1
    if k==1:
        if np.all( vp[-y-b:-y ,x-a:x+a] == bg_clr) :
            up(vp,a,b,x,y,rgb,i,n)
            j+=1
        m+=1
    if k==2:
        if np.all( vp[x:x+b, y-a:y+a] == bg_clr) :
            down(vp,a,b,x,y,rgb,i,n)
            j+=1
        m+=1 
    if k==3:
        if np.all( vp[y-a:y+a , -x-b:-x ] == bg_clr) :
            left(vp,a,b,x,y,rgb,i,n)
            j+=1
        m+=1
    return j, m,rgb
def contur(k,j,m,vp,a,b,x,y,bg_clr,i,n):
    rgb = (0,0,0)
    if k==1:
        if np.all( vp[-y-b-20:-y ,x-a:x+a] == bg_clr) :
            while(n<=b):
                w=i+1
                c=w*w
    #             vp[-y-c:-y-n ,x-i-2:x+i+2 ] = color 
                vp[-y-c:-y-n ,x-i-1:x-i ] = rgb
                vp[-y-c:-y-n ,x+i:x+i+1 ] = rgb
                i+=1
                n=i*i
            vp[-y-n-1:-y-n ,x-i:x+i ] = rgb
            j+=1
        m+=1
#     вниз
    if k==2:
        if np.all( vp[y:y+b+20 ,x-a:x+a] == bg_clr) :
            while(n<=b):
                w=i+1
                c=w*w
    #             vp[y+n:y+c ,x-i-2:x+i+2 ] = color
                vp[y+n:y+c ,x-i-1:x-i ] = rgb
                vp[y+n:y+c ,x+i:x+i+1 ] = rgb
                i+=1
                n=i*i
            vp[y+n:y+n+1 ,x-i:x+i ] = rgb
            j+=1
        m+=1  
# влево
    if k==3:
        if np.all( vp[y-a-1:y+a+1 , -x-b:-x] == bg_clr):
            c = 1
            while(i<=a and -x-c > -x-b):
                w=i+1
                c=w*w
    #             vp[y-i:y+i ,-x-b:-x-c] =color
                if -x-c < -x-b:
                    vp[y-i-1:y-i ,-x-b:-x-n ] = rgb
                    vp[y+i:y+i+1 ,-x-b:-x-n ] = rgb
                    n = b
                else:
                    vp[y-i-1:y-i ,-x-c:-x-n ] = rgb
                    vp[y+i:y+i+1 ,-x-c:-x-n ] = rgb
                    i+=1
                    n=i*i
            vp[y-i:y+i ,-x-n:-x-n+1 ] = rgb
            j+=1
        m+=1
# вправо
    if k==0:
        if np.all( vp[y-a-1:y+a+1 , x:x+b] == bg_clr):
            c = 1
            while(i<=a and x+c < x+b):
                w=i+1
                c=w*w
    #             vp[y-i:y+i ,x+c:x+b] =color
                if x+c > x+b:
                    vp[y-i-1:y-i ,x+n:x+b ] = rgb
                    vp[y+i:y+i+1 ,x+n:x+b ] = rgb
                    n = b
                else:
                    vp[y-i-1:y-i ,x+n:x+c] = rgb
                    vp[y+i:y+i+1 ,x+n:x+c ] = rgb
                    i+=1
                    n=i*i
            vp[y-i:y+i ,x+n-1:x+n ] = rgb
            j+=1
        m+=1
    return j,m,rgb

def Mix(k,j,m,vp,a,b,x,y,bg_clr,i,n):
    color = np.random. randint(0,255, size  =3)
    rgb = (0,0,0)
    if k==0:
        if np.all( vp[-y-b:-y ,x-a:x+a] == bg_clr) :
            while(n<=b):
                w=i+1
                c=w*w
                vp[max(-y-c,-y-b):-y-n ,x-i:x+i ] = color 
                vp[max(-y-c,-y-b):-y-n ,x-i-1:x-i ] = rgb
                vp[max(-y-c,-y-b):-y-n ,x+i:x+i+1 ] = rgb
                i+=1
                n=i*i
            vp[-y-b-1:-y-b ,x-i:x+i ] = rgb
            j+=1
        m+=1
#     вниз
    if k==1:
        if np.all( vp[y:y+b ,x-a:x+a] == bg_clr) :
            c= 1
            while(c<b):
                w=i+1
                c=w*w
                vp[y+n:min(y+c,y+b) ,x-i:x+i ] = color
                vp[y+n:min(y+c,y+b) ,x-i-1:x-i ] = rgb
                vp[y+n:min(y+c,y+b) ,x+i:x+i+1 ] = rgb
                i+=1
                n=i*i
            vp[y+b-1:y+b ,x-i:x+i ] = rgb
            j+=1
        m+=1 
# влево
    if k==2:
        if np.all( vp[y-a-1:y+a+1 , -x-b:-x] == bg_clr):
            c = 1
            while(i<=a and -x-c > -x-b):
                w=i+1
                c=w*w
                vp[y-i-1:y+i+1 ,-x-b:-x-c] =color
                if -x-c < -x-b:
                    vp[y-i-1:y-i ,-x-b:-x-n ] = rgb
                    vp[y+i:y+i+1 ,-x-b:-x-n ] = rgb
                    n = b
                else:
                    vp[y-i-1:y-i ,-x-c:-x-n ] = rgb
                    vp[y+i:y+i+1 ,-x-c:-x-n ] = rgb
                    i+=1
                    n=i*i
            vp[y-i:y+i ,-x-n:-x-n+1 ] = rgb
            j+=1
        m+=1
# вправо
    if k==3:
        if np.all( vp[y-a-1:y+a+1 , x:x+b] == bg_clr):
            c = 1
            while(i<=a and x+c < x+b):
                w=i+1
                c=w*w
                vp[y-i-1:y+i+1 ,x+c:x+b] =color
                if x+c > x+b:
                    vp[y-i-1:y-i ,x+n:x+b ] = rgb
                    vp[y+i:y+i+1 ,x+n:x+b ] = rgb
                    n = b
                else:
                    vp[y-i-1:y-i ,x+n:x+c] = rgb
                    vp[y+i:y+i+1 ,x+n:x+c ] = rgb
                    i+=1
                    n=i*i
            vp[y-i:y+i ,x+n-1:x+n ] = rgb
            j+=1
        m+=1
    return j,m,color
```

**14. Однородные координаты. Целесообразность употребления в компьютерной графике.**



Однородными координатами точки (x, y) - это тройка x1, x2, x3, таких, что:

$$ \frac{x1}{x3} = x ; \frac{x2}{x3} = y $$

При решении задач компьютерной графики однородные координаты обычно вводятся следующим образом:
точке **М(х, у)** плоскости ставится в соответствие точка **М*(х, у, 1)**.
В общем случаем точке **(х, у)** ставится в соответствие точка **(hx, hy, h) при h ≠ 0**.

Матрицы поворота, масштабирования, зеркального отражения и переноса в однородных координатах:

$$

0 & 0 & 1
\end{bmatrix}
$$

0 & 0 & 1
\end{bmatrix}
$$

$$

### 15. Растровая и векторная графика.
Растровая и векторная графика представляют собой два основных подхода к созданию и обработке изображений. Каждый из них имеет свои особенности, преимущества и недостатки.

**Растровая графика**

    Определение:
        Растровая графика, или пиксельная графика, представляет изображение в виде сетки пикселей. Каждый пиксель имеет определённый цвет и значение, а изображение формируется путем объединения этих пикселей.

    Характеристики:

        1. Разрешение: Изображения имеют фиксированное разрешение, которое определяется числом пикселей в ширину и высоту.
    
        2. Размер файла: Размер файла может быть большим, особенно для изображений с высоким разрешением.
        
        3. Качество: При увеличении размера растрового изображения может наблюдаться потеря качества (пикселизация).
    
        4. Форматы: Основные форматы включают JPEG, PNG, GIF, BMP, TIFF.
    
    Примеры:
        * Фотографии
        * Скриншоты
        * Детализированные изображения с множеством цветовых оттенков

    Использование:
        * Растровая графика идеальна для отображения сложных и детализированных изображений, таких как фотографии и рисунки, где важны детали и оттенки.
    
**Векторная графика**
    *Определение:
        * Векторная графика использует математические формулы для определения геометрических форм, таких как линии, кривые и многоугольники. Изображение формируется из этих объектов, которые описываются параметрами.
        
    Характеристики:
        1. Разрешение: Векторные изображения масштабируемы без потери качества. Их можно увеличивать или уменьшать до любого размера без искажения.
        2. Размер файла: Обычно размер файла меньше по сравнению с растровыми изображениями, так как хранится информация о математических формулах, а не о каждом пикселе.
        3. Качество: Качество не зависит от размера изображения. Объекты остаются четкими и резкими при любом увеличении.
        4. Форматы: Основные форматы включают SVG, EPS, AI, PDF.

    Примеры:
        * Логотипы
        * Иконки
        * Иллюстрации и схемы
    
    Использование:
        * Векторная графика лучше подходит для создания масштабируемых изображений, таких как логотипы и иконки, которые могут использоваться в различных размерах.


### 16. Полигональная модель.

**Полигональная модель** — это способ представления трехмерных объектов в компьютерной графике, где объект описывается с помощью полигонов, в основном треугольников и многоугольников. Полигональные модели являются основным способом представления трехмерных объектов в современных системах моделирования и рендеринга.

Основные компоненты полигональной модели

**Вершины (Vertices):**
 *Это точки в трехмерном пространстве, которые определяют углы полигонов. Вершины задаются с помощью координат 
(
𝑥
,
𝑦
,
𝑧
)
(x,y,z).

**Ребра (Edges):**
    *Это линии, соединяющие пары вершин. Ребра образуют границы полигонов.
    
**Полигоны (Polygons):**
    *Многоугольники, образуемые соединением трех или более вершин. 
    *Наиболее часто используемые полигоны — треугольники и четырехугольники.
    
**Нормали (Normals):**
    *Векторы, перпендикулярные поверхности полигонов. Нормали используются для определения освещенности и создания реалистичных эффектов освещения.
    
##### **Создание и обработка полигональных моделей**

**Создание моделей:**

**Моделирование:**
    Используются различные методы моделирования, такие как полигональное моделирование, субдивизионное моделирование и другие. В таких методах можно вручную редактировать вершины, ребра и полигоны для создания сложных объектов.
    
**Генерация:**
Модели могут быть созданы автоматически с помощью алгоритмов генерации, таких как процедурное моделирование.

**Сглаживание:**
    *Сглаживание вершин: Для устранения острых углов и создания более плавных переходов между полигонами.
    *Сглаживание полигонов: Используется для улучшения визуального качества модели, сглаживая переходы между полигонами.
    
##### **Текстурирование:**

**UV-развёртка:**
    *Процесс развертывания поверхности 3D-модели на 2D-плоскость, чтобы применить текстуру. Это позволяет наложить изображение текстуры на модель. 
    
**Tекстуры:** Изображения, применяемые к поверхности модели для добавления деталей и цвета.
    
**Освещение и рендеринг:**
    *Освещение: Нормали используются для расчета освещенности модели, что влияет на её визуальное восприятие.
    *Рендеринг: Преобразование 3D-модели в 2D-изображение. Включает расчет освещения, теней, текстур и других визуальных эффектов.
    
**Примеры использования полигональных моделей**
    *Компьютерные игры: Полигональные модели используются для создания персонажей, окружающего мира и объектов.
    *Анимация: В мультфильмах и анимациях полигональные модели могут быть анимированы и модифицированы для создания движения.
    *Продуктовый дизайн: Для создания и визуализации продуктов перед их производством.
    *Визуализация архитектуры: Создание моделей зданий и интерьеров для предварительного просмотра и проектирования.
