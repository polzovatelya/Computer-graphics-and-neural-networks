```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
W=H=400
bg_clr = (255,255,255)
vp = np.full((H, W, 3), bg_clr, dtype = 'uint8')

vp[:, :] = (255,255,255) #изменение всех элементов массива в одной строчке


# границы 2
width = 5
blue = (0,0,255)
vp[:width,:] = blue
vp[:,-width:] = blue
vp[-width:,:] = blue
vp[:,:width] = blue
```

```python
def contur(k,j,m,vp,a,b,x,y,bg_clr,i,n):
    rgb = np.random. randint(10,120, size  =3)
    if k==1:
        if np.all( vp[-y-b-20:-y ,x-a:x+a] == bg_clr) :
            while(n<=b):
                w=i+1
                c=w*w
    #             vp[-y-c:-y-n ,x-i-2:x+i+2 ] = color 
                vp[-y-c:-y-n ,x-i-1:x-i ] = rgb
                vp[-y-c:-y-n ,x+i:x+i+1 ] = rgb
                i+=1
                n=i*i
            vp[-y-n-1:-y-n ,x-i:x+i ] = rgb
            j+=1
        m+=1
#     вниз
    if k==2:
        if np.all( vp[y:y+b+20 ,x-a:x+a] == bg_clr) :
            while(n<=b):
                w=i+1
                c=w*w
    #             vp[y+n:y+c ,x-i-2:x+i+2 ] = color
                vp[y+n:y+c ,x-i-1:x-i ] = rgb
                vp[y+n:y+c ,x+i:x+i+1 ] = rgb
                i+=1
                n=i*i
            vp[y+n:y+n+1 ,x-i:x+i ] = rgb
            j+=1
        m+=1  
# влево
    if k==3:
        if np.all( vp[y-a-1:y+a+1 , -x-b:-x] == bg_clr):
            c = 1
            while(i<=a and -x-c > -x-b):
                w=i+1
                c=w*w
    #             vp[y-i:y+i ,-x-b:-x-c] =color
                if -x-c < -x-b:
                    vp[y-i-1:y-i ,-x-b:-x-n ] = rgb
                    vp[y+i:y+i+1 ,-x-b:-x-n ] = rgb
                    n = b
                else:
                    vp[y-i-1:y-i ,-x-c:-x-n ] = rgb
                    vp[y+i:y+i+1 ,-x-c:-x-n ] = rgb
                    i+=1
                    n=i*i
            vp[y-i:y+i ,-x-n:-x-n+1 ] = rgb
            j+=1
        m+=1
# вправо
    if k==0:
        if np.all( vp[y-a-1:y+a+1 , x:x+b] == bg_clr):
            c = 1
            while(i<=a and x+c < x+b):
                w=i+1
                c=w*w
    #             vp[y-i:y+i ,x+c:x+b] =color
                if x+c > x+b:
                    vp[y-i-1:y-i ,x+n:x+b ] = rgb
                    vp[y+i:y+i+1 ,x+n:x+b ] = rgb
                    n = b
                else:
                    vp[y-i-1:y-i ,x+n:x+c] = rgb
                    vp[y+i:y+i+1 ,x+n:x+c ] = rgb
                    i+=1
                    n=i*i
            vp[y-i:y+i ,x+n-1:x+n ] = rgb
            j+=1
        m+=1
    return j,m,rgb
```

```python
from PIL import Image

j = 0
m = 0
N = 20
k=0
p = randint(0,2) 
while  j < N and m < 1000  :
    a,b = np.random. randint(10,120, size  =2)
    y= randint(width+a, H-a - width-2) # физический у 
    x= randint(width+a, W-b - width-2)
    k = randint(0,3) 
    i=0
    n=0
    j,m,color = contur(k,j,m,vp,a,b,x,y,bg_clr,i,n)
    
    

        
plt.axis('off')
plt.imshow(vp)
plt.show()
```
